```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
```

```python
data = []
with open("../log.txt", "r") as f:
    lines = f.readlines()
    for line in lines:
        data.append(float(line.strip()))
```

```python
sns.histplot(data=data, kde=True, bins=25)
plt.xlabel("Total response time")
plt.title("Total Response Time Distribution (Concurrent Number=10)")
plt.savefig("time_dist.jpg", dpi=300)
plt.savefig("time_dist.pdf", format="pdf", bbox_inches="tight")
```

```python
sns.boxplot(x=data, palette="muted")

plt.title('Box plot of Total Response Time (Concurrent Number = 10)')
plt.xlabel('Total response time')
plt.savefig("box.jpg", dpi=300)
plt.savefig("box.pdf", format="pdf", bbox_inches="tight")
```

```python
st.norm.interval(alpha=0.95,
                 loc=np.mean(data),
                 scale=st.sem(data))
```
